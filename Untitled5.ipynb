```python
def sinplt():
    x=np.linspace(0,14,100)
    for i in range (1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i))
sb.set_style('darkgrid')
sinplt()
plt.show()
```

```python
cur=sb.color_palette("Greens")
sb.palplot(cur)
plt.show()
```

```python
d=sb.load_dataset('iris')
sb.distplot(d['petal_length'])
plt.show()
```

```python
d=sb.load_dataset('iris')
sb.jointplot(x='petal_length',y='petal_width',data=d)
plt.show()
print(d)
```

```python
d=sb.load_dataset('iris')
sb.jointplot(x='petal_length',y='petal_width',data=d,kind='kde')
plt.show()
```

```python
d=sb.load_dataset('iris')
sb.set_style('ticks')
sb.pairplot(d,hue='species',diag_kind='kde',kind='reg',palette='husl')
plt.show()
```

```python
d=sb.load_dataset('iris')
sb.stripplot(x='species',y='petal_length',data=d,jitter=True)
plt.show()
```

```python
d=sb.load_dataset('iris')
sb.boxplot(x='species',y='petal_length',data=d)
plt.show()
```
