```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
featur_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in featur_na:
    print(feature, np.round(dataset[feature].isnull().mean(),2),'% missing values')
```

```python
for feature in featur_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtype != 'O']
print('Numerical variable is ' ,len(numerical_feature))
dataset[numerical_feature].head()
```

```python
feature_year=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
feature_year
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
feature_year
```

```python
 for feature in feature_year:
        if feature != 'YrSold':
            data=dataset.copy()
            data[feature]=data['YrSold']-data[feature]
            plt.scatter(data[feature],data['SalePrice'])
            plt.show()
```
