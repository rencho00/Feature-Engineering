```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,24,2)
y=np.arange(10,130,10)
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')
plt.show()
```

```python
x=np.arange(0,3.14*2,0.05)
y=np.sin(x)
fig=plt.figure()
a=fig.add_axes([0,0,1,1])
a.plot(x,y)
a.set_title("sine Wave")
a.set_xlabel("angle")
a.set_ylabel('sine')
a.legend(labels=('TV','sdda'),loc='upper right')
plt.show()
plt.plot([1,2,3])
plt.plot(range(12))
```

```python
a=np.random.random_integers(10,100,200)
b=np.random.random_integers(30,80,200)
c=np.random.random_integers(20,90,200)
d=np.random.random_integers(15,70,200)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
data_to_plot=[a,b,c,d]
bp=ax.boxplot(data_to_plot)
plt.show()
```
