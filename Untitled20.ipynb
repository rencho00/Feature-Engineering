```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=sb.load_dataset('titanic')
data.head()
```

```python
data.isnull().sum()
```

```python
data['age']=data['age']
random=data['age'].dropna().sample(data['age'].isnull().sum(),random_state=0)
random.index=data[data['age'].isnull()].index
data.loc[data['age'].isnull(),'age']=random
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['embarked'])
```

```python
data['deck']=data['deck'].astype('string')
data['deck']=data['deck'].fillna("MISSING")
data['deck']=data['deck'].astype('category')
data['deck']=data['deck'].astype(str).str[0]
```

```python
data.isnull().sum()
```

```python
data=data.drop(columns=['embark_town'])
```

```python
data.isnull().sum()
```

```python
sb.heatmap(data.isnull(),yticklabels=False)
```

```python
data.head()
```

```python
data['sex']=pd.get_dummies(data['sex'])
```

```python
data.head()
```

```python
mea=data.groupby(['deck'])['survived'].mean().index
```

```python
mea=mea.sort_values()
```

```python
mea
```

```python
dic ={k:i for i,k in enumerate(mea,0)}
```

```python
dic
```

```python
data['deck']=data['deck'].map(dic)
```

```python
data
```

```python
data['class']=pd.get_dummies(data['class'])
```

```python
data
```

```python
data=data.drop(columns=['fare','who','alone','adult_male','alive'])
```

```python
data
```

```python
X=data.drop(columns=['survived'])
y=data['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
X_train=s.fit_transform(X_train)
X_test=s.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lmode=LogisticRegression()
```

```python
lmode.fit(X_train,y_train)
pred=lmode.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
