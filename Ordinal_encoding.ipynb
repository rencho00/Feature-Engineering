# Count Encoding

```python
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,index_col=None)
```

```python
data.head()
```

```python
colums=[1,3,5,6,7,8,9,13]
```

```python
data=data[colums]
```

```python
data.head()
```

```python
data.columns=['Employ','Degree','Status','Designation','relation','race','sex','country']
```

```python
data.head()
```

```python
for col in data.columns:
    print(col,"     ",len(data[col].unique()), 'values')
```

```python
di=data['country'].value_counts().to_dict()
```

```python
data['country']=data['country'].map(di)
```

# Target Guided Encoding

```python
df=pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
colums=['cabin','survived']
```

```python
df=df[colums]
```

```python
df.head()
```

```python
df['cabin']=df['cabin'].fillna('Missing')
```

```python
df['cabin']=df['cabin'].astype(str).str[0]
```

```python
df.groupby(['cabin'])['survived'].mean()
```

```python
df.groupby(['cabin'])['survived'].mean().sort_values()
```

```python
ordinal= df.groupby(['cabin'])['survived'].mean().sort_values().index
```

```python
ordinal
```

```python
label= {k:i for i,k in enumerate(ordinal,0)}
label
```

```python
df['rank']=df['cabin'].map(label)
```

```python
df.drop(columns='ran')
```

# MEAN ENCODING

```python
mean_ordinal=df.groupby(['cabin'])['survived'].mean().to_dict()
```

```python
df['mean_ordinal']=df['cabin'].map(mean_ordinal)
```

# Probability encoding

```python
import pandas as pd
import numpy as np
```

```python
ds=pd.read_csv('titanic_train.csv')
ds.head()
```

```python
columns=['survived','cabin']
ds=ds[columns]
ds['cabin']=ds['cabin'].fillna("Missing")
```

```python
ds['cabin']=ds['cabin'].astype(str).str[0]
```

```python
ds.head(20)
```

```python
pf=ds.groupby(['cabin'])['survived'].mean()
```

```python
pf=pd.DataFrame(pf)
pf['died']=1-pf['survived']
```

```python
pf['prob_ratio']=pf['survived']/pf['died']
```

```python
pf    
```

```python
encode=pf['prob_ratio'].to_dict()
```

```python
ds['prob_encode']=ds['cabin'].map(encode)
```
