```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score,silhouette_samples 
```

```python
X,y=make_blobs(n_samples = 500, n_features =2, centers = 4, random_state =1)
k=range(2,10)
mean =[]
for i in k:
    km= KMeans(n_clusters=i, random_state=3)
    km.fit(X)
    mean.append(sum(np.min(cdist(X,km.cluster_centers_,'euclidean'),axis=1))/X.shape[0])
    print(silhouette_score(X,km.fit_predict(X)))
plt.plot(k,mean,'bx-')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python

X,y = make_blobs(n_samples = 500, n_features =2,centers = 4,random_state = 10)
k= range(1,10)
for n in k: 
    for i in range(1,n):  
        ag=AgglomerativeClustering(n_clusters =10,affinity = 'euclidean',linkage='ward')
        y_hc = ag.fit_predict(X)
        plt.scatter(X[y_hc==i,0], X[y_hc ==i,1], s= 50 )
    
    plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendo = sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
db=DBSCAN(eps=0.5,min_samples =10)
yp = db.fit_predict(X)
labels = db.labels_
plt.scatter(X[:,0],X[:,1],c=yp)
print("No of Clusters is : {}".format(len(set(yp[np.where(yp!=-1)]))))
print("Silhouette Score :{}".format(metrics.silhouette_score(X,labels)))
```
